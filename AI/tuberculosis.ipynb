```python
import os
base_dir = '/content/drive/MyDrive/data'

import pandas as pd
filepaths, labels = [], []
for cls in ['Normal', 'Tuberculosis']:
    cls_dir = os.path.join(base_dir, cls)
    for fname in os.listdir(cls_dir):
        if fname.lower().endswith(('.png','.jpg','.jpeg')):
            filepaths.append(os.path.join(cls_dir, fname))
            labels.append(cls)
df = pd.DataFrame({'filename': filepaths, 'class': labels})

from sklearn.model_selection import train_test_split
train_df, temp_df = train_test_split(df, test_size=0.30, stratify=df['class'], random_state=42)
valid_df, test_df = train_test_split(temp_df, test_size=0.50, stratify=temp_df['class'], random_state=42)
print(f"Train: {len(train_df)} | Val: {len(valid_df)} | Test: {len(test_df)}")

import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator

batch_size = 16
img_height = 500
img_width  = 500

datagen_train = ImageDataGenerator(rescale=1./255, horizontal_flip=True)
datagen_test  = ImageDataGenerator(rescale=1./255)

train_gen = datagen_train.flow_from_dataframe(
    train_df, x_col='filename', y_col='class',
    target_size=(img_height, img_width), color_mode='grayscale',
    class_mode='binary', batch_size=batch_size
)

valid_gen = datagen_test.flow_from_dataframe(
    valid_df, x_col='filename', y_col='class',
    target_size=(img_height, img_width), color_mode='grayscale',
    class_mode='binary', batch_size=batch_size, shuffle=False
)

test_gen = datagen_test.flow_from_dataframe(
    test_df, x_col='filename', y_col='class',
    target_size=(img_height, img_width), color_mode='grayscale',
    class_mode='binary', batch_size=batch_size, shuffle=False
)

plt.figure(figsize=(12, 6))
for images, labels in train_gen:
    for i in range(6):
        plt.subplot(2, 3, i+1)
        plt.imshow(images[i].squeeze(), cmap='gray')
        plt.title('Tuberculosis' if labels[i] == 1 else 'Normal')
        plt.axis('off')
    break
plt.show()

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.utils.class_weight import compute_class_weight

classes_array = train_gen.classes
unique_classes = np.unique(classes_array)
weights = compute_class_weight('balanced', classes=unique_classes, y=classes_array)
cw = dict(zip(unique_classes, weights))
print("Class weights:", cw)

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(img_height, img_width, 1)),
    MaxPooling2D(2,2),
    Conv2D(32,(3,3),activation='relu'), MaxPooling2D(2,2),
    Conv2D(64,(3,3),activation='relu'), MaxPooling2D(2,2),
    Conv2D(64,(3,3),activation='relu'), MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'), Dropout(0.5),
    Dense(64, activation='relu'),  Dropout(0.5),
    Dense(1, activation='sigmoid'),
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

callbacks = [
    EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True),
    ReduceLROnPlateau(monitor='val_loss', patience=2, factor=0.3, min_lr=1e-6, verbose=1)
]

history = model.fit(
    train_gen,
    epochs=25,
    validation_data=valid_gen,
    class_weight=cw,
    callbacks=callbacks
)

import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

pd.DataFrame(history.history)[['loss','val_loss']].plot(title='Loss'); plt.show()
pd.DataFrame(history.history)[['accuracy','val_accuracy']].plot(title='Accuracy'); plt.show()

test_loss, test_acc = model.evaluate(test_gen)
print(f"Test accuracy: {test_acc*100:.2f}%")

preds = (model.predict(test_gen) > 0.5).astype(int).ravel()
y_true = test_gen.classes
cm = confusion_matrix(y_true, preds)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=['Normal','Tuberculosis'], yticklabels=['Normal','Tuberculosis'])
plt.xlabel('Predicted'); plt.ylabel('Actual'); plt.show()
print(classification_report(y_true, preds, target_names=['Normal','Tuberculosis']))

from tensorflow.keras.preprocessing import image

def predict_tuberculosis(img_path, model, target_size=(img_height, img_width)):
    img = image.load_img(img_path, target_size=target_size, color_mode='grayscale')
    arr = image.img_to_array(img) / 255.0
    arr = np.expand_dims(arr, axis=0)
    prob = model.predict(arr)[0][0]
    label = 'Tuberculosis' if prob > 0.5 else 'Normal'
    return label, prob

```

```python
label, prob = predict_tuberculosis(
    '/content/drive/MyDrive/data/Normal/Normal-1005.png', model)
print(f'Prediction: {label} ({prob*100:.2f}% confidence)')
```
